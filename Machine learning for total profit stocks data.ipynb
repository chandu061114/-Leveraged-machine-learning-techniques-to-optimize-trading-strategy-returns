```python
def calculate_additional_features(data):
    # Relative Strength Index (RSI)
    
    # Bollinger Bands
    window = 20
    data['ma'] = data['Close'].rolling(window=window).mean()
    data['std'] = data['Close'].rolling(window=window).std()
    data['upper_band'] = data['ma'] + (2 * data['std'])
    data['lower_band'] = data['ma'] - (2 * data['std'])

    # Moving Averages (5, 20, 50, 100)
    data['ma_5'] = data['Close'].rolling(window=5).mean()
    data['ma_20'] = data['Close'].rolling(window=20).mean()
    data['ma_50'] = data['Close'].rolling(window=50).mean()
    data['ma_100'] = data['Close'].rolling(window=100).mean()

    # Two Volume-based Indicators
    data['volume_ma_5'] = data['Volume'].rolling(window=5).mean()
    data['volume_ma_20'] = data['Volume'].rolling(window=20).mean()

    # Percentage Change for Last 1, 3, 5 Days
    data['pct_change_1d'] = data['Close'].pct_change(1)
    data['pct_change_3d'] = data['Close'].pct_change(3)
    data['pct_change_5d'] = data['Close'].pct_change(5)

    # High and Low Prices for Last 1, 3, 5, 8, 13, 21 Days
    periods = [1, 3, 5, 8, 13, 21]
    for period in periods:
        data[f'high_{period}d'] = data['High'].rolling(window=period).max()
        data[f'low_{period}d'] = data['Low'].rolling(window=period).min()

    # Peaks and Troughs
    data['peak'] = (data['Close'] > data['Close'].shift(1)) & (data['Close'] > data['Close'].shift(-1))
    data['trough'] = (data['Close'] < data['Close'].shift(1)) & (data['Close'] < data['Close'].shift(-1))

    return data.fillna(0)

# Apply the function to the existing DataFrame
daily_data = calculate_additional_features(daily_data)

daily_data
```

```python
def calculate_additional_features(data, stocks):
    for stock in stocks:
        # Define the close column for the current stock
        close_col = f'{stock}_close'
        
        
        # Bollinger Bands for the stock
        window = 20
        data[f'{stock}_ma'] = data[close_col].rolling(window=window).mean()
        std = data[close_col].rolling(window=window).std()
        data[f'{stock}_upper_band'] = data[f'{stock}_ma'] + (2 * std)
        data[f'{stock}_lower_band'] = data[f'{stock}_ma'] - (2 * std)

        # Moving Averages for the stock
        for ma in [5, 20, 50, 100]:
            data[f'{stock}_ma_{ma}'] = data[close_col].rolling(window=ma).mean()

        # Volume-based Indicators - Adjust if volume data for each stock is available
        # If stock-specific volume data isn't available, consider removing these lines or adjust accordingly
        # data[f'{stock}_volume_ma_5'] = data['Volume'].rolling(window=5).mean()
        # data[f'{stock}_volume_ma_20'] = data['Volume'].rolling(window=20).mean()

        # Percentage Change for Last 1, 3, 5 Days for the stock
        data[f'{stock}_pct_change_1d'] = data[close_col].pct_change(1)
        data[f'{stock}_pct_change_3d'] = data[close_col].pct_change(3)
        data[f'{stock}_pct_change_5d'] = data[close_col].pct_change(5)

        # High and Low Prices for Last 1, 3, 5, 8, 13, 21 Days for the stock
        # Adjust if high/low data for each stock is available or use nifty_data's High/Low as approximation
        periods = [1, 3, 5, 8, 13, 21]
        for period in periods:
            data[f'{stock}_high_{period}d'] = data['High'].rolling(window=period).max() # Adjust if stock-specific High data available
            data[f'{stock}_low_{period}d'] = data['Low'].rolling(window=period).min() # Adjust if stock-specific Low data available

        # Peaks and Troughs for the stock - Adjust if needed based on stock-specific Close data
        data[f'{stock}_peak'] = (data[close_col] > data[close_col].shift(1)) & (data[close_col] > data[close_col].shift(-1))
        data[f'{stock}_trough'] = (data[close_col] < data[close_col].shift(1)) & (data[close_col] < data[close_col].shift(-1))

    return data.fillna(0)

# Define the list of stocks (based on the columns you've added to `nifty_data` for closing prices)
stocks = ['hdfcbank', 'nsei','icicibank', 'kotakbank', 'sbin', 'axisbank', 'indusindbk']

# Apply the function to the existing DataFrame
daily_data = calculate_additional_features(daily_data, stocks)

daily_data
```

```python
import numpy as np
import pandas as pd

# Assuming daily_data is your DataFrame and 'total_profit' is the column of interest
# Define the range of p values you want to test
p_values = np.arange(0, daily_data['total_profit'].max(), 10)  # Adjust the step as needed

# Initialize variables to store the best p value and the smallest class difference
best_p = None
smallest_diff = float('inf')

# Iterate over each potential p value
for p in p_values:
    # Create the binary target variable for the current p
    daily_data['is_profitable'] = np.where(daily_data['total_profit'] > p, 1, 0)
    # Get the class distribution
    class_distribution = daily_data['is_profitable'].value_counts()
    # Calculate the difference between classes
    class_diff = abs(class_distribution[0] - class_distribution[1])
    # If the current class difference is smaller than the smallest found so far, update best_p and smallest_diff
    if class_diff < smallest_diff:
        best_p = p
        smallest_diff = class_diff

# Output the best p value
print(f"The best p value is: {best_p} with a class difference of: {smallest_diff}")

# Apply the best p value to create the binary target variable
daily_data['is_profitable'] = np.where(daily_data['total_profit'] > 80, 1, 0)
# Recalculate the value_counts_df with the best p value
value_counts_series = daily_data['is_profitable'].value_counts()
value_counts_df = value_counts_series.reset_index()
value_counts_df.columns = ['Value', 'Count']

# Display the class distribution with the best p
print(value_counts_df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import numpy as np
from sklearn.linear_model import LogisticRegression

# Selecting features and the new target
X = daily_data.drop(['total_profit', 'long_trades', 'short_trades', 'number_of_trades', 'is_profitable', 'cumulative_profit', 'long_profit', 'short_profit'], axis=1)  # Assuming all other columns are features
y = daily_data['is_profitable']

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False, test_size=0.3, random_state=42)

# Scaling features (important for models like Logistic Regression)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize and train the Logistic Regression model with L2 regularization
model = LogisticRegression(C=1.0, penalty='l2', solver='liblinear')  # Adjust C for regularization strength
model.fit(X_train_scaled, y_train)
```

```python

# Evaluate the model
predictions = model.predict(X_test_scaled)
prediction_probs = model.predict_proba(X_test_scaled)[:, 1]

y_pred = model.predict(X_test_scaled)
accuracy = accuracy_score(y_test, predictions)
conf_matrix = confusion_matrix(y_test, predictions)
class_report = classification_report(y_test, predictions)

print(f"Accuracy: {accuracy}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(class_report)

```

```python
# Assuming 'combined_results' is your DataFrame with actual and predicted values
# And 'daily_data' contains the 'total_profit' for each day

# Add the 'total_profit' column to the combined_results DataFrame
combined_results['total_profit'] = daily_data.loc[combined_results.index, 'total_profit']

# Calculate modified profit based on the model's predictions
combined_results['modified_profit'] = combined_results['Predicted'] * combined_results['total_profit']

# Calculate cumulative sums for actual and modified profits
combined_results['cumulative_profit'] = combined_results['total_profit'].cumsum()
combined_results['cumulative_modified_profit'] = combined_results['modified_profit'].cumsum()
```

```python
# Assuming 'Actual' reflects whether the day was truly profitable (1) or not (0) based on some criteria
# Note: If 'Actual' is not in the expected format, adjust this step accordingly to reflect actual profits

# Determine the profit/loss outcome based on 'modified_profit' and 'High_Prob_Profit'
combined_results['modified_profit_outcome'] = (combined_results['modified_profit'] > 0).astype(int)
combined_results['high_prob_profit_outcome'] = (combined_results['High_Prob_Profit'] > 0).astype(int)
```

```python
# Calculate accuracy for modified profits
correct_modified_predictions = (combined_results['modified_profit_outcome'] == combined_results['Actual']).sum()
total_predictions = len(combined_results)
modified_profit_accuracy = correct_modified_predictions / total_predictions

# Calculate accuracy for high-probability profits
# Filtering out days without high probability predictions
high_prob_predictions_exist = combined_results['High_Prob_Profit'] != 0
correct_high_prob_predictions = (combined_results.loc[high_prob_predictions_exist, 'high_prob_profit_outcome'] == combined_results.loc[high_prob_predictions_exist, 'Actual']).sum()
total_high_prob_predictions = high_prob_predictions_exist.sum()
high_prob_profit_accuracy = correct_high_prob_predictions / total_high_prob_predictions

# Calculate normal accuracy by comparing 'Predicted' directly with 'Actual'
correct_predictions = (combined_results['Predicted'] == combined_results['Actual']).sum()
total_predictions = len(combined_results)
normal_accuracy = correct_predictions / total_predictions


print(f"Normal Accuracy: {normal_accuracy:.2f}")
print(f"Modified Profit Accuracy: {modified_profit_accuracy:.2f}")
print(f"High Probability Profit Accuracy: {high_prob_profit_accuracy:.2f}")
```

```python
# Assuming 'combined_results' contains your dataset
import pandas as pd

# Initialize summary DataFrame structure
summary_data = {
    'Type': ['Normal', 'Model-Based', 'High Probability'],
    'Total Days': [len(combined_results)] * 3,
    'Profitable Days': [None] * 3,  # Placeholder for calculations
    'Loss Days': [None] * 3,  # Placeholder for calculations
    'Non-Trading Days': [0, 0, 0],  # Initialized with 0; will be updated for Model-Based and High Probability
    'Accuracy': [None] * 3,  # Placeholder for new accuracy calculations
    'Frequency': [None] * 3,  # Placeholder for frequency calculations
    'Total Profits': [0] * 3,  # Placeholder for profit calculations
    'Total Losses': [0] * 3,  # Placeholder for loss calculations
    'Average Profit': [0] * 3,  # Placeholder for average profit calculations
    'Average Loss': [0] * 3,  # Placeholder for average loss calculations
}

# Create the DataFrame
summary_df = pd.DataFrame(summary_data)

# Calculations for each scenario
for i, scenario in enumerate(['Normal', 'Model-Based', 'High Probability']):
    if scenario == 'Normal':
        condition = combined_results['total_profit'] != 0  # All days are considered for Normal
    elif scenario == 'Model-Based':
        condition = combined_results['modified_profit'] != 0
    else:  # High Probability
        condition = combined_results['High_Prob_Profit'] != 0

    trading_data = combined_results[condition]
    trading_days = len(trading_data)
    profitable_days = trading_data[trading_data['total_profit'] > 0].shape[0]
    loss_days = trading_data[trading_data['total_profit'] <= 0].shape[0]
    total_profits = trading_data[trading_data['total_profit'] > 0]['total_profit'].sum()
    total_losses = trading_data[trading_data['total_profit'] <= 0]['total_profit'].sum()
    avg_profit = trading_data[trading_data['total_profit'] > 0]['total_profit'].mean()
    avg_loss = trading_data[trading_data['total_profit'] <= 0]['total_profit'].mean()
    non_trading_days = summary_df.loc[i, 'Total Days'] - trading_days

    # Update DataFrame
    summary_df.loc[i, 'Profitable Days'] = profitable_days
    summary_df.loc[i, 'Loss Days'] = loss_days
    summary_df.loc[i, 'Non-Trading Days'] = non_trading_days
    summary_df.loc[i, 'Total Profits'] = total_profits
    summary_df.loc[i, 'Total Losses'] = total_losses
    summary_df.loc[i, 'Average Profit'] = avg_profit
    summary_df.loc[i, 'Average Loss'] = avg_loss
    summary_df.loc[i, 'Accuracy'] = profitable_days / trading_days if trading_days > 0 else 0
    summary_df.loc[i, 'Frequency'] = trading_days / summary_df.loc[i, 'Total Days']

```
